```python
import pandas as pd
```

```python
df = pd.read_json('DPLabels_2020.json')
df.index.rename('Index', inplace = True)
df = df.reset_index()

vardf = pd.DataFrame()
for i in range(len(df)):
    var = df.iloc[i]['variables']
    varSer = pd.Series(var)
    vardf = vardf.append(varSer, ignore_index=True)
vardf['Metric'] = df['Index']

dp20 = vardf[vardf['group'].isin(['DP02', 'DP03', 'DP04', 'DP05'])]
dp20 = dp20[['group', 'Metric', 'label']]
dp20
```
